```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/processed/absenteeism_cleaned.csv')
print(f"Loaded cleaned data: {df.shape}")
df.head()
```

```python
# Create absence categories
# Group absenteeism hours into categories for additional analysis
df['Absence_Category'] = pd.cut(df['Absenteeism time in hours'], 
                                 bins=[0, 4, 8, 24, 120],
                                 labels=['Short', 'Half_Day', 'Full_Day', 'Extended'])

print("Absence Category Distribution:")
print(df['Absence_Category'].value_counts())
```

```python
# Create season features
season_mapping = {1: 'Summer', 2: 'Autumn', 3: 'Winter', 4: 'Spring'}
df['Season_Name'] = df['Seasons'].map(season_mapping)

print("Season Distribution:")
print(df['Season_Name'].value_counts())
```

```python
# Create workload categories
df['Workload_Category'] = pd.cut(df['Work load Average/day'], 
                                  bins=[0, 250, 300, 400],
                                  labels=['Low', 'Medium', 'High'])

print("Workload Category Distribution:")
print(df['Workload_Category'].value_counts())

print("\nFeature engineering complete!")
df.head()
```
